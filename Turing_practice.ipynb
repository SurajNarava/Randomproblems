**1 Valid parantheses**
# image.png

```python
def isValid(s: str) -> bool:
    openstr = ['(','{','[']
    closestr = [')','}',']']
    lastchar = []
    for char in s:
        if char in openstr:
            lastchar.append(char)
        elif char in closestr:
            if len(lastchar) == 0:
                return False
            if char == closestr[openstr.index(lastchar[-1])]:
                lastchar.pop(-1)
            else:
                return False
    return len(lastchar) == 0


line = input()
if isValid(line):
    print("valid")
else:
    print("invalid")
```
